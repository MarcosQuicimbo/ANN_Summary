```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python

# Generar datos de muestra
X, y = make_regression(n_samples=200, n_features=1, noise=0.1, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# Crear y entrenar el modelo
# MLP regressor is created with a single hidden layer containing 50 neurons
mlp_regressor = MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000, random_state=1)
mlp_regressor.fit(X_train, y_train)

# Realizar predicciones
y_pred = mlp_regressor.predict(X_test)

# Visualizar los resultados
plt.scatter(X_test, y_test, label='True values')
plt.scatter(X_test, y_pred, label='Predictions', marker='x')
plt.legend()
plt.show()

# Puntuación del modelo
score = mlp_regressor.score(X_test, y_test)
#que tal buena es la prediccion del modelo frente al valor real 
print("R^2 Score:", score)
```
